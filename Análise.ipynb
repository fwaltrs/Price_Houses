# Boxplots das variáveis contínuas

```python
import matplotlib.pyplot as plt
import seaborn as sns

continuous_vars =  ['price', 'bedrooms', 'bathrooms', 'sqft_living',
'sqft_lot', 'floors']


# Criar os boxplots
plt.figure(figsize=(30, 20))
for i, var in enumerate(continuous_vars, 1):
    plt.subplot(2, 3, i)  # Alterado para 2 linhas e 3 colunas
    sns.boxplot(y=base[var], color="purple")
    plt.title(f"{var}", fontsize=25)  # Aumenta o tamanho do título
    plt.xlabel(f"{var}", fontsize=20)  # Aumenta o tamanho do rótulo do eixo X
    plt.ylabel("Value", fontsize=20)  # Aumenta o tamanho do rótulo do eixo Y
    plt.xticks(fontsize=15)  # Aumenta o tamanho dos rótulos do eixo X
    plt.yticks(fontsize=15)  # Aumenta o tamanho dos rótulos do eixo Y
    plt.tight_layout()

plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

continuous_vars = [
    "price",
    "bedrooms",
    "bathrooms",
    "sqft_living",
    "sqft_lot",
    "floors",
    #"waterfront",
    #"view",
    "grade",
    "sqft_above",
    "sqft_basement",
    "lat",
    "long",
    "sqft_living15",
    "sqft_lot15"
]

plt.figure(figsize=(20, 15))
for i, var in enumerate(continuous_vars, 1):
    plt.subplot(5, 3, i)  # Organizado em uma grade de 5x3
    sns.histplot(base[var], kde=False, color="purple")  # Inclui uma curva KDE
    plt.title(f"{var}")
    plt.tight_layout()

plt.show()
```

```python
corr = base.select_dtypes(include=['int', 'float']).corr()

# Criar o heatmap com ajustes
plt.figure(figsize=(12, 8))  # Ajuste o tamanho do gráfico
sns.heatmap(corr, cmap='OrRd', annot=True, fmt='.2f', annot_kws={"size": 10}, linewidths=0.5)

# Adicionar título
plt.title('Correlação', fontsize=16)

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Calcular a correlação entre as variáveis numéricas
corr = base.select_dtypes(include=['int', 'float']).corr()

# Criar uma máscara para a parte superior da matriz (remover valores repetidos)
mask = np.triu(np.ones_like(corr, dtype=bool))  # Máscara para a parte superior

# Criar o heatmap com a máscara
plt.figure(figsize=(12, 8))  # Ajuste o tamanho do gráfico
sns.heatmap(corr, cmap='OrRd', annot=True, fmt='.2f', annot_kws={"size": 10},
            linewidths=0.5, mask=mask, cbar_kws={'label': 'Correlação'})

# Adicionar título
plt.title('Correlação', fontsize=16)

# Mostrar o gráfico
plt.show()
```

```python
from collections import Counter
# Contando a frequência de cada valor
contagem = Counter(base['price'])

# Determinando a moda
frequencia_maxima = max(contagem.values())
modas = [k for k, v in contagem.items() if v == frequencia_maxima]

# Exibindo as modas e suas frequências
for moda in modas:
    print(f'A moda é {moda}, e aparece {contagem[moda]} vezes.')

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Linear regression example
x_linear = np.linspace(0, 10, 100)
y_linear = 0.15 * x_linear - 0.1

# Logistic regression example
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x_logistic = np.linspace(-10, 10, 100)
y_logistic = sigmoid(x_logistic)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Linear Regression
axes[0].plot(x_linear, y_linear, color='purple', label='Regressão Linear')
axes[0].scatter(x_linear, [0 if i < 5 else 1 for i in x_linear], color='black', edgecolors='black')
axes[0].set_title('Regressão Linear')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Probabilidade')
axes[0].set_xlim(0, 10)
axes[0].set_ylim(-0.1, 1.1)
axes[0].axhline(1, linestyle='--', color='gray')
axes[0].grid(True, linestyle='--', alpha=0.6)
# axes[0].legend()

# Logistic Regression
axes[1].plot(x_logistic, y_logistic, color='purple', label='Regressão Logística')
axes[1].scatter(x_logistic, [0 if i < 0 else 1 for i in x_logistic], color='black', edgecolors='black')
axes[1].set_title('Regressão Logística')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Probabilidade')
axes[1].set_xlim(-10, 10)
axes[1].set_ylim(-0.1, 1.1)
axes[1].axhline(1, linestyle='--', color='gray')
axes[1].axhline(0, linestyle='--', color='gray')
axes[1].grid(True, linestyle='--', alpha=0.6)
# axes[1].legend()

plt.tight_layout()
plt.show()
```
